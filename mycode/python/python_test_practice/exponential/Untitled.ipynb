```python
def expo(x):
    import numpy as np
    import matplotlib.pyplot as plt
    kmax = 200
    a = x.size
    abcisa = np.linspace(0,a-1,a)
    i_range = x.size
    exp = list(x)
    exp_val = 0
    fac = 1
    exp_num = np.exp(x)
    tol = 1e-14
    for i in range(i_range):
        exp_val = 0
        for j in range(kmax):
            exp_0 = exp_val
            if j ==0:
                fac = 1
            elif j>0:
                fac = j*fac
            if i == 0:
                exp_val = 1
            elif i > 0:
                exp_val = ((x[i])**j/fac) + exp_val
                delta = exp_val - exp_0
                if (abs(delta/x[i])) < tol:
                    break
            exp[i] = exp_val
    #print(exp)
    #print(exp_num)
    figure, axis = plt.subplots(2)
    axis[0].plot(abcisa,exp)
    axis[1].plot(abcisa,exp_num)
    plt.show()
```

```python
import numpy as np
x = np.linspace(0,70,71)
expo(x)
%timeit expo(x)
```
